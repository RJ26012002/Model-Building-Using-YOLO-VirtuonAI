```python
def convert_to_yolo(json_file_path, output_file_path,h,w):
    class_mapping = {
        "TABLE": 0,
        "BUYER": 1,
        "DATE": 2,
        "NUMBER": 3,
        "SELLER_ADDRESS": 4,
        "SELLER_NAME": 5,
        "SUB_TOTAL": 6,
        "TITLE": 7,
        "TOTAL": 8,
        "GSTIN": 9,
        "GST(7%)": 10,
        "LOGO": 11,
        "DISCOUNT" : 12,
        "DUE_DATE" : 13,
        "GSTIN_BUYER" : 14,
        "NOTE" : 15,
        "PAYMENT_DETAILS" : 16,
        "PO_NUMBER" : 17,
        "SELLER_EMAIL" : 18,
        "SELLER_SITE" : 19,
        "TAX" : 20,
        "TOTAL_WORDS" : 21,
        "GSTIN_SELLER" : 22,
        "BILL_TO" : 23,
        "GST(9%)" : 24,
        "SEND_TO" : 25,
        "CONDITIONS" : 26,
        "AMOUNT_DUE" : 27,
        "GST(1%)" : 28,
        "GST(18%)" : 29 

    }
    c= 0
    with open(json_file_path) as f:
        data = json.load(f)
    # print(data)
    with open(output_file_path, 'w') as output_file:
        for category, category_data in data.items():
            # print("the category is ",category , " data is : ",category_data) 
            
            if category in class_mapping:
                
                class_index = class_mapping[category]
                # print("the class_index is ",category)
                if isinstance(category_data, list):
                    # print("The count is c=",c)
                    c=c+1
                    # print("THe kist items are ",list)
                    # print("the item is " , category_data , "for category ", category)
                    for item in category_data:
                        # print("the item is " , item)
                        
                        dict = item[0]
                        # print("the dict is ",dict)
                        x = dict.keys()
                        if 'bbox' in dict:
                            x = dict['bbox']
                            if len(x) == 2:
                                a = x[0]
                                x1=a[0]
                                y1 =a[1]
                                b=x[1]
                                x2 = b[0]
                                y2 = b[1]
                                # if x1==0 or y1==0 or x2 == 0 or y2 == 0:
                                #     continue
                                # if x1>x2 or y1>y2:
                                #     continue
                                y1 = 841 - y1
                                y2 = 841 - y2
                                p = min(x1,x2)
                                q = min(y1,y2)
                                r = max(x1,x2)
                                s = max(y1,y2)
                                # line1 = f"{class_index} {x1} {y1} {x2} {y2}\n"
                                # line = f"{class_index} {(x1+x2)/(2*w)} {(y1+y2)/(2*h)} {(x2-x1)/w} {(y1-y2)/h}\n"
                                voc_bbox = (p,q,r,s)
                                # voc_bbox = (x1 , y1,x2,y2)
                                print("the pqrs is ",voc_bbox)
                                W, H = w, h  # WxH of the image
                                c = pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(W,H))
                                a = round(c[0],6)
                                b= round(c[1],6)
                                d= round(c[2],6)
                                e= round(c[3],6)

                    

                                # line10=  f"{class_index} {c[0]} {c[1]} {c[2]} {c[3]}\n"
                                line = f"{class_index} {a} {b} {d} {e}\n"
                                # line = f"{class_index} {c[0]} {c[1]} {c[2]} {c[3]}\n"
                                # print("before ",line1)
                                # print("after ",line)
                                print("the text int line is " , line )
                                output_file.write(line)
                else:
                    dict = category_data
                    x=dict['bbox']
                    a = x[0]
                    x1=a[0]
                    y1 =a[1]
                    b=x[1]
                    x2 = b[0]
                    y2 = b[1]
                    # if x1==0 or y1==0 or x2 == 0 or y2 == 0:
                    #     continue
                    # if x1>x2 or y1>y2:
                    #     continue
                    # print(x1,y1,x2,y2)
                    y1 = 841 - y1
                    y2 = 841 - y2
                    p = min(x1,x2)
                    q = min(y1,y2)
                    r = max(x1,x2)
                    s = max(y1,y2)
                    # print(p,q,r,s)
                    # line1 = f"{class_index} {x1} {y1} {x2} {y2}\n"
                    # print( "the main isssssssssssssssssss ",x1+x2 , 2*w)
                    # line = f"{class_index} {(x1+x2)/(2*w)} {(y1+y2)/(2*h)} {(x2-x1)/w} {(y2-y1)/h}\n"
                    voc_bbox = (p,q,r,s)
                    # voc_bbox = (2.0 ,617.4058000000001, 185.09599999999998, 701.4058)
                    # print("the type of voc bbox is "),
                    # print(type(voc_bbox))
                    # print("the type of x1 is ")
                    # print(type(x1))
                    # print("the voc_bbox is ",voc_bbox)
                    W, H = w, h  # WxH of the image
                    # print("the width and height is ",W,H)
                    c = pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(W,H))
                    # print("the c output is ",c)
                    a = round(c[0],6)
                    b= round(c[1],6)
                    d= round(c[2],6)
                    e= round(c[3],6)

                    

                    # line10=  f"{class_index} {c[0]} {c[1]} {c[2]} {c[3]}\n"
                    line = f"{class_index} {a} {b} {d} {e}\n"
                   
                    # print("before ",line1)
                    # print("line 3 is " , line3)
                    # print("line10 : ",line10) 
                    print("line : ",line)
                    output_file.write(line)
                    # print("the data is " , category_data)
                    # print(type(category_data))
                    # print("the text int line is " , line )

# Example usage:
path = 'D:/YOLO_Test/valid/json_labels' # the path from where you have to pick your files
myPicList = os.listdir(path)
dest_path = 'D:/YOLO_Test/valid/txt_labels'  # the path from where you have to save your files 

for j,y in enumerate(myPicList):
    json_file_path = path+'/'+y
    original_name = y[:-5]
    output_file_path = dest_path+'/'+original_name+'.txt'
    w = 595
    h = 841
    convert_to_yolo(json_file_path, output_file_path,h,w)
    # print("Done for ",y)
print("succesffully done")

# json_file_path = "Template10_Instance4.json"
# output_file_path = "Template10_Instance4.txt"
# img = Image.open("Template9_Instance6.jpg")
# # # height, width = img.size
# w= int(img.size[0])
# h= int(img.size[1])
# # print("Height =", h)
# # print("Width =", w)
# convert_to_yolo(json_file_path, output_file_path,h,w)
# # original_name = json_file_path[:-5]
# # # print(path)
# # output_file_path = dest_path+'/'+original_name+'.txt'
# # # print(output_file_path)
# # convert_to_yolo(json_file_path, output_file_path)
```
