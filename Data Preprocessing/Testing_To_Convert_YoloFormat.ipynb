```python
img = Image.open("Template9_Instance6.jpg")

def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

# im=Image.open(img_path)
w= int(img.size[0])
h= int(img.size[1])
print("w",w)
print("h",h)

print( 90.0 ,708.4058 ,241.092 ,720.4058 ) #define your x,y coordinates 90.0 708.4058 241.092 720.4058
b = (90.0 ,708.4058 ,241.092 ,720.4058 )
bb = convert((w,h), b)
print(bb)
```

```python
def convert_bbox(x1, y1, x2, y2, image_width, image_height):
    X = (x1 + x2) / 2
    Y = (y1 + y2) / 2
    W = x2 - x1
    H = y2 - y1

    print("the H we are getting is ",H)
    X_normalized = X / image_width
    Y_normalized = Y / image_height
    W_normalized = W / image_width
    H_normalized = H / image_height

    return X_normalized, Y_normalized, W_normalized, H_normalized

# Example usage:
x1, y1, x2, y2 =2.0, 398.8898, 542.0, 326.8898
image_width, image_height = 595, 841
yolo_bbox = convert_bbox(x1, y1, x2, y2, image_width, image_height)
print(yolo_bbox)  # Output: (0.15, 0.15, 0.1, 0.1) 0.15714285714285714, 0.7835909631391201, 0.30756302521008405, 0.09988109393579073
```

```python
def bbox_to_yolo(x1, y1, x2, y2, image_width, image_height):
    X = (x1 + x2) / 2
    Y = (y1 + y2) / 2
    W = x2 - x1
    H = y2 - y1

    u = X / image_width
    v = Y / image_height
    w = W / image_width
    h = H / image_height

    return u, v, w, h
print(bbox_to_yolo(2.0 ,617.4058000000001, 185.09599999999998, 701.4058,595,841)) # 0.15714285714285714, 0.7835909631391201, 0.30756302521008405, 0.09988109393579073
```
